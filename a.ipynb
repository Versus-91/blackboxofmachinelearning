```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, axs = plt.subplots(2)

# Plot data on subplots
axs[0].plot(x, y1, label='sin(x)')
axs[1].plot(x, y2, label='cos(x)')

# Collect handles and labels from all subplots
handles_dict = {}
for ax in axs:
    handles_ax, labels_ax = ax.get_legend_handles_labels()
    for handle, label in zip(handles_ax, labels_ax):
        if label not in handles_dict:
            handles_dict[label] = handle

# Create a single legend for all subplots
unique_handles = list(handles_dict.values())
unique_labels = list(handles_dict.keys())
fig.legend(unique_handles, unique_labels, loc='upper right')

plt.show()
```
